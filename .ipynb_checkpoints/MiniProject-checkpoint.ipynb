# Creating Decision Tree Model

```python
#import encoder from sklearn
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
```

```python
# OneHotEncoding of categorical predictors
data_cat = data[['gender', 'platform', 'profession', 'interests']]
ohe.fit(data_cat)
data_cat_ohe = pd.DataFrame(ohe.transform(data_cat).toarray(),
                           columns=ohe.get_feature_names_out(data_cat.columns))

# Check the encoded variables
data_cat_ohe.info()
```

```python
# Combining Numeric features with OHE categorical features
data_num = data[['age', 'income']]
data_res = data['time_spent_category']
data_ohe = pd.concat([data_num, data_cat_ohe, data_res], sort = False, axis = 1).reindex(index=data_num.index)

# Check the final dataframe
data_ohe.info()
```

# Balancing classes of time_spent_category

```python
count_high, count_low = data['time_spent_category'].value_counts()
print("Ratio of classes is High : Low = ", count_high, ":", count_low)
```

```python
# Upsample Low to match High
from sklearn.utils import resample

time_high = data_ohe[data_ohe.time_spent_category == 'high']
time_low = data_ohe[data_ohe.time_spent_category == 'low']
 
# Upsample the Bad samples
time_low_up = resample(time_low, 
                        replace=True,                     # sample with replacement
                        n_samples=time_high.shape[0])    # to match number of High
```

```python
# Combine the two classes back after upsampling
data_ohe_up = pd.concat([time_high, time_low_up])
 
# Check the ratio of the classes
data_ohe_up['time_spent_category'].value_counts()
```

```python
# Quick plot to check the balanced classes visually
sb.catplot(y = 'time_spent_category', data = data_ohe_up, kind = "count")
```

```python
# Confirm that the OHE is still in place
# and that the samples have now increased
data_ohe_up.info()
```

# Creating and Fitting the Model

```python
# Import essential models and functions from sklearn
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.tree import plot_tree
```

```python
# Extract Response and Predictors
y = pd.DataFrame(data_ohe_up['time_spent_category'])
x = pd.DataFrame(data_ohe_up.drop('time_spent_category', axis = 1))
                 
# Split the Dataset into Train and Test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
# Decision Tree using Train Data
dectree = DecisionTreeClassifier(max_depth = 4) 

# train the decision tree model
dectree.fit(x_train, y_train)                    

# Plot the trained Decision Tree
f = plt.figure(figsize=(24,24))

plot_tree(dectree, filled=True, rounded=True, 
          feature_names=x_train.columns.tolist(), 
          class_names=["low","high"])
```

# Checking accuracy of model on Train data

```python
# Predict the Response corresponding to Predictors
y_train_pred = dectree.predict(x_train)

# Print the Classification Accuracy
print("Train Data")
print("Accuracy  :\t", dectree.score(x_train, y_train))
print()

# Print the Accuracy Measures from the Confusion Matrix
cmTrain = confusion_matrix(y_train, y_train_pred)
tpTrain = cmTrain[1][1] # True Positives : Good (1) predicted Good (1)
fpTrain = cmTrain[0][1] # False Positives : Bad (0) predicted Good (1)
tnTrain = cmTrain[0][0] # True Negatives : Bad (0) predicted Bad (0)
fnTrain = cmTrain[1][0] # False Negatives : Good (1) predicted Bad (0)

print("TPR Train :\t", (tpTrain/(tpTrain + fnTrain)))
print("TNR Train :\t", (tnTrain/(tnTrain + fpTrain)))
print()

print("FPR Train :\t", (fpTrain/(tnTrain + fpTrain)))
print("FNR Train :\t", (fnTrain/(tpTrain + fnTrain)))

# Plot the two-way Confusion Matrix
sb.heatmap(confusion_matrix(y_train, y_train_pred), 
           annot = True, fmt=".0f", annot_kws={"size": 18})
```

# Checking accuracy of model on Test data

```python
# Import the required metric from sklearn
from sklearn.metrics import confusion_matrix

# Predict the Response corresponding to Predictors
y_test_pred = dectree.predict(x_test)

# Print the Classification Accuracy
print("Test Data")
print("Accuracy  :\t", dectree.score(x_test, y_test))
print()

# Print the Accuracy Measures from the Confusion Matrix
cmTest = confusion_matrix(y_test, y_test_pred)
tpTest = cmTest[1][1] # True Positives : Good (1) predicted Good (1)
fpTest = cmTest[0][1] # False Positives : Bad (0) predicted Good (1)
tnTest = cmTest[0][0] # True Negatives : Bad (0) predicted Bad (0)
fnTest = cmTest[1][0] # False Negatives : Good (1) predicted Bad (0)

print("TPR Test :\t", (tpTest/(tpTest + fnTest)))
print("TNR Test :\t", (tnTest/(tnTest + fpTest)))
print()

print("FPR Test :\t", (fpTest/(fpTest + tnTest)))
print("FNR Test :\t", (fnTest/(fnTest + tpTest)))

# Plot the two-way Confusion Matrix
sb.heatmap(confusion_matrix(y_test, y_test_pred), 
           annot = True, fmt=".0f", annot_kws={"size": 18})
```
